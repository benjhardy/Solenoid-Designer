```python
import numpy as np
from math import pi
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 25})
plt.rcParams.update({'font.weight': 'bold'})
```

```python
def getFreq(L,C):
    f = 1/(2*pi*np.sqrt(np.multiply(L,C)))
    return f

def getC(a,b):
    k = 2
    c = (.241*k)/(np.log10(b/a)) # pF/cm
    return c/1e12

def getL(a,b):
    l = .00459*np.log10(b/a) # uH/cm
    return l/1e6
```

```python
# a is smaller radius
a = 5.7 #mm
b = np.linspace(a + .1,10,50) #mm
print(b.shape)

leng = 1 # cm
cs = getC(a,b)*leng
ls = getL(a,b)*leng

freqs = getFreq(ls,cs)/1e6 # GHz

plt.figure(figsize=(30,10))

plt.subplot(1,3,1)
plt.plot(b,cs)
plt.ylabel("C/cm (Farads)")
plt.xlabel('outer radius (mm)')

plt.subplot(1,3,2)
plt.plot(b,ls)
plt.ylabel("L/cm (Henries)")
plt.xlabel('outer radius (mm)')

plt.subplot(1,3,3)
plt.plot(b,freqs)
plt.ylabel("freq. (MHz)")
plt.xlabel('outer radius (mm)')
plt.show()


l = np.linspace(4,7,50) # 1 - 7cm 
cs = np.multiply(getC(a,b),l)
ls = np.multiply(getL(a,b),l)
freqs = getFreq(ls,cs)/1e6 # MHz
plt.plot(l,freqs)
plt.ylabel("freq. (MHz)")
plt.xlabel('length (cm)')
plt.title('Self-resonance of two cylinders vs length')
plt.show()
```
